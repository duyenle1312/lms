```python
import re

keywords_dict = {}
stop_words = ["a", "an", "the", "and", "but", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "can", "will", "just"]

for index, i in clean_courses.iterrows():
    dept = i["Keyword"]
    keyword = i["Skill gain"]

    arr = keyword.split(',')
    for i in range(len(arr)):
        word = re.sub('\W+',' ', arr[i].lower()).strip()
        if word not in stop_words:
            arr[i] = word

    if dept in keywords_dict:
        #print("a: ", arr)
        keywords_dict[dept] += arr
        
    else:
        #print("b: ", arr)
        keywords_dict[dept] = arr
   

# print(keywords_dict)
for i, j in keywords_dict.items():
    new_keyword = set(j)
    new_keyword= list(new_keyword)
    new_keyword.sort()
    keywords_dict[i] = new_keyword
    # print(i, j)

print(keywords_dict)
```

```python
import pandas as pd

keywords_data = pd.read_csv("data.csv")
new_data = keywords_data.dropna(subset=["Keywords"])
new_data.to_csv("course_with_keywords.csv", sep='\t', encoding='utf-8', index=False, header=True)
new_data
```

```python
keywords_list = []
for index, i in new_data.iterrows():
    keyword = i["Keywords"]

    arr = keyword.split(',')
    keywords_list.extend(arr)
    #print(keywords_list)

keywords_set = list(set(keywords_list))
keywords_set.sort()
print(len(keywords_set))
#keywords_set
```

```python
# initialize list of lists
columns = ["ID", "Title"] + keywords_set

result = {}
# Create the pandas DataFrame
for col in columns:
    result[col] = []

id = 1
for index, item in new_data.iterrows():
    keyword_items = item["Keywords"].split(",")

    for i in keywords_set:

        if i in keyword_items:
            result[i].append(1)
        else:
            result[i].append(0)

    result["ID"].append(id)
    id += 1
    result["Title"].append(item["Course Title"])

#print(result)
df = pd.DataFrame(result, columns=columns)

df
```

```python
df.to_csv("course_keyword.csv", sep='\t', encoding='utf-8', index=False, header=True)
```

```python
# initialize list of lists
user_columns = ["User ID"] + keywords_set

user_result = {}
user_result["User ID"] = [0]
# Create the pandas DataFrame
user_interests = { "Artificial Intelligence": 10, "Race & Racism": 9, 
                  "Data Science": 8, "Technology": 7, "Software Development": 6,
                  "Web Development": 5, "Cloud Technology": 4, "Business": 3, "Branding": 2, "Design": 1}

for col in keywords_set:
    if col in user_interests:
        user_result[col] = [user_interests[col]]
    else:
        user_result[col] = [0]

print(user_result)

user_df = pd.DataFrame(user_result, columns=user_columns)

user_df
```
