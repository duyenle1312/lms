```python
import pandas as pd
```

```python
courses = pd.read_csv("CourseraDataset.csv")
```

```python
eng_courses = courses[courses['Course Title'].map(lambda x: x.isascii())]
eng_courses
```

```python
clean_courses = eng_courses.dropna(how='any',axis=0) 
clean_courses = clean_courses.loc[clean_courses['What you will learn'] != "Not specified"]
clean_courses
```

```python
clean_courses = clean_courses.drop_duplicates(subset=['Course Title'])
clean_courses
```

```python
clean_courses.to_csv("course_data.csv", sep='\t', encoding='utf-8', index=False, header=True)
```
